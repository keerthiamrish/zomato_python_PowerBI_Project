```python
#Removing the columns with entire null values
df.drop(columns=['Unnamed: 10','Unnamed: 11','Unnamed: 12','Unnamed: 13'],inplace=True)
df.head()
```

The columns were dropped because they contained no meaningful data (entirely null values) and therefore do not contribute to the analysis.

```python
(df.isnull().sum() / len(df)) * 100
```

Checking missing values in percentage helps in better understanding the data and making informed decisions during analysis.

```python
#***********DINING RATING************#
#Cleaning and filling the appropriate values
print("Dining Rating Datatype:",df['Dining Rating'].dtype)

#Checking whether to treat mean or median
print('Mean values:',df['Dining Rating'].mean())
print('Median Value',df['Dining Rating'].median())

#Checking the distribution
sns.boxplot(x=df['Dining Rating'])
Q1 = df['Dining Rating'].quantile(0.25)
Q3 = df['Dining Rating'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Dining Rating'] < lower_bound) | (df['Dining Rating'] > upper_bound)]

outliers.shape
```

The distribution is slightly left-skewed as the mean is lower than the median, indicating the presence of lower-end outliers influencing the data.Since the data distribution is skewed and contains outliers, the median was used to fill missing values in the Dining Rating column to ensure robust analysis.

```python
#Filling Median For Null Values
df['Dining Rating'].fillna(df['Dining Rating'].median(), inplace=True)
print("Dining Rating Datatype:",df['Dining Rating'].dtype)
df = df[df['Dining Rating'] > 0]
print("Dining Rating contains Null values:",df['Dining Rating'].isnull().sum())
```

Missing values in Dining Rating were filled using the median to maintain consistency and avoid data loss.

```python
#**************DELIVERY RATING***************#
#Cleaning and filling the appropriate values
print("Delivery Rating Datatype:",df['Delivery Rating'].dtype)

#Checking whether to treat mean or median
print('Mean values:',df['Delivery Rating'].mean())
print('Median Value',df['Delivery Rating'].median())

#Checking the distribution
sns.boxplot(x=df['Delivery Rating'])
Q1 = df['Delivery Rating'].quantile(0.25)
Q3 = df['Delivery Rating'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Delivery Rating'] < lower_bound) | (df['Delivery Rating'] > upper_bound)]

outliers.shape
```

The Delivery Rating distribution shows slight left skewness with the presence of outliers. Hence, the median was used to fill missing values to ensure robustness against extreme values.

```python
#Filling Median For Null Values
df['Delivery Rating'].fillna(df['Delivery Rating'].median(), inplace=True)
print("Delivery Rating Datatype:",df['Delivery Rating'].dtype)
print("Deliverry Rating contains Null values:",df['Delivery Rating'].isnull().sum())
```

The Delivery Rating column exhibited slight left skewness with the presence of outliers. Therefore, missing values were imputed using the median to ensure robustness. Post-imputation, the column was verified to contain no null values and maintained the correct float datatype.

```python
#********** DINING RATING COUNT***********#
#Cleaning and filling the null values
df['Dining Rating Count'] = df['Dining Rating Count'].replace(['Does not offering dining','Not enough Dining Reviews','None',''], 0)
```

```python
df['Dining Rating Count'] = pd.to_numeric(df['Dining Rating Count'], errors='coerce')
print("Dining rating Datatype:",df['Dining Rating Count'])
```

```python

df['Dining Rating Count'].fillna(df['Dining Rating Count'].median(),inplace=True)
print("Dining rating Datatype:",df['Dining Rating Count'])
```

```python
df['Dining Rating Count']=df['Dining Rating Count'].astype(int)
print("Dining rating Datatype:",df['Dining Rating Count'])
```

Text values such as ‘No Dining’ were replaced with 0, followed by numeric conversion, median imputation for missing values, and final conversion to integer datatype.

```python
#*****************DELIVERY RATING COUNT**************#
#Cleaning and filling the null values
#Replace text values with 0
df['Delivery Rating Count'] = df['Delivery Rating Count'].replace(
    ['Does not offer Delivery', 'Not enough Delivery Reviews', None],
    0
)

#Convert to numeric
df['Delivery Rating Count'] = pd.to_numeric(df['Delivery Rating Count'], errors='coerce')

#Fill any remaining nulls with median
df['Delivery Rating Count'].fillna(df['Delivery Rating Count'].median(), inplace=True)

#Convert to integer
df['Delivery Rating Count'] = df['Delivery Rating Count'].astype(int)

#Check
print(df['Delivery Rating Count'].dtype)
print("Delivery Rating Count null values:",df['Delivery Rating Count'].isnull().sum())
```

The Delivery Rating Count column contained non-numeric values such as "Does not offer Delivery" and "Not enough Delivery Reviews", which were replaced with 0 to indicate absence of delivery activity. The column was then converted to numeric format, missing values were handled using median, and finally converted to integer datatype.

```python
df['Features'].value_counts().head(10)
```

```python
#******************FEATURES COLUMN****************#
#Home Delivery
df['Home Delivery'] = df['Features'].apply(lambda x: 1 if 'Home Delivery' in str(x) else 0)

#Indoor Seating
df['Indoor Seating'] = df['Features'].apply(lambda x: 1 if 'Indoor Seating' in str(x) else 0)

#Delivery Only
df['Delivery Only'] = df['Features'].apply(lambda x: 1 if 'Delivery Only' in str(x) else 0)
df.head()
```

The Features column was transformed into binary variables such as Home Delivery, Indoor Seating, and Delivery Only, where 1 indicates presence of the feature and 0 indicates absence.

```python
#********PRICE CATEGORY******#
df['Price Category'] = pd.cut(df['Price for 2'],bins=[0, 300, 700, 1500, 3000],labels=['Low', 'Medium', 'High', 'Luxury'])
df.head()
```

The pd.cut function was used to categorize the continuous ‘Price for 2’ column into meaningful groups such as Low, Medium, High, and Luxury, making the data easier to analyze and interpret.

```python
#**************RATING CATEGORY*************#
df['Rating Category'] = pd.cut(df['Dining Rating'], bins=[0,2.5, 3.5, 4.5, 5], labels=['Poor', 'Average', 'Good', 'Excellent'])
df.head()
```

The Dining Rating column was categorized into qualitative groups such as Poor, Average, Good, and Excellent using binning techniques to simplify analysis and interpretation

```python
#*****************POPULARITY SCORE*************#
df['Popularity Score'] = df['Dining Rating'] * df['Dining Rating Count']
df.head()
```

A Popularity Score was created by combining dining rating and rating count to identify restaurants that are both highly rated and widely reviewed, providing a more reliable measure of popularity.

```python
#***********FEATURE COUNT************#
df['Feature Count'] = df['Features'].apply(lambda x: len(eval(x)) if isinstance(x, str) else 0)
df.head()
```

A Feature Count variable was created by converting the Features column into a list format and counting the number of services offered by each restaurant, providing a numerical representation of service availability.

```python
#*************MAIN CUISINE********#
df['Main Cuisine'] = df['Cuisine'].apply(lambda x: x.split(',')[0].strip() if isinstance(x, str) else x)
df.head()
```

The Main Cuisine feature was created by extracting the primary cuisine from the Cuisine column to simplify analysis and enable better categorization of restaurants

```python
#************RATING DIFFERENCE**********#
df['Rating Difference'] = df['Dining Rating'] - df['Delivery Rating']
df.head()
```

Rating Difference feature was created to compare dining and delivery performance, helping identify whether restaurants perform better in dine-in or delivery services.

```python
#************VALUE SCORE**************#
df['Value Score'] = df['Dining Rating'] / df['Price for 2']*100
df.head()
```

A Value Score (Cost Efficiency) metric was created to evaluate the quality of restaurants relative to their pricing

```python
#*****************HIGH RATING****************#
df['High Rated'] = df['Dining Rating'].apply(lambda x: 1 if x >= 4 else 0)
df.head()
```

A High Rated indicator was created to classify restaurants with a rating of 4 and above, enabling easier comparison between well-rated and average restaurants.

```python
df.info()
```

```python
#Price Category
# Convert to object (remove category restriction)
df['Price Category'] = df['Price Category'].astype('object')

# Fill missing values
df['Price Category'].fillna('Unknown', inplace=True)

# Convert back to category (optional)
df['Price Category'] = df['Price Category'].astype('category')
df['Price Category'].isnull().sum()
```

Missing values in the Price Category column were handled by converting the column to object type, imputing ‘Unknown’, and converting it back to categorical format for efficient analysis

```python
#Rating Category
# Convert to object
df['Rating Category'] = df['Rating Category'].astype('object')

# Fill missing values
df['Rating Category'].fillna('Unknown', inplace=True)

# Convert back to category
df['Rating Category'] = df['Rating Category'].astype('category')
df['Rating Category'].isnull().sum()
```

Missing values in the Rating Category column were handled by assigning ‘Unknown’ after temporarily converting the column to object type.

```python
#Dropping Top Dishes
df.drop(columns=['Top Dishes'],inplace=True)
df.head()
```

The Top Dishes column was removed as it contained unstructured text data and was not relevant for the analysis

```python
#Removing Duplicate
#Checking Duplicate values
df.duplicated().sum()

#Removing Duplicate
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

Duplicate records were identified and removed to ensure data consistency and avoid redundancy in analysis.

```python
df.info()
```

The dataset contains 12,027 records and 20 columns with no missing values. All features were cleaned and transformed with appropriate data types, including numerical, categorical, and engineered variables, making the dataset fully prepared for analysis
